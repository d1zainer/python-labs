**Задача 1**

Реализовать сортировку массива целых чисел по неубыванию их модуля.

Запрещено использовать:
sorted(),
list.sort(),
numpy и т.д.

Необходимо реализовать алгоритм сортировки вручную (сортировка вставками или выбором). Провести тестирование минимум на 3 наборах данных, включая крайние случаи.

Используется сортировка вставками.

Алгоритм проходит по массиву слева направо и на каждом шаге вставляет текущий элемент в уже отсортированную часть массива, сравнивая элементы по их модулю.


Сравнение выполняется через abs(), так как требуется сортировка по модулю числа.

```python
def sort_by_absolute(arr):
    n = len(arr)
    
    for i in range(1, n):
        current = arr[i]
        j = i - 1
        
        # Сдвигаем элементы вправо,
        # пока модуль предыдущего больше текущего
        while j >= 0 and abs(arr[j]) > abs(current):
            arr[j + 1] = arr[j]
            j -= 1
        
        arr[j + 1] = current
    
    return arr
```

```python
import random

# 1. Пустой массив
test1 = []
print("Тест 1:", sort_by_absolute(test1.copy()))

# 2. Один элемент
test2 = [5]
print("Тест 2:", sort_by_absolute(test2.copy()))

# 3. Отрицательные и положительные числа
test3 = [3, -1, -7, 4, 2, -2]
print("Тест 3:", sort_by_absolute(test3.copy()))

# 4. Случайный массив
test4 = [random.randint(-50, 50) for _ in range(20)]
print("Исходный:", test4)
print("Отсортированный:", sort_by_absolute(test4.copy()))
```

В алгоритме используется два вложенных цикла:

Внешний цикл выполняется n раз.

Внутренний цикл в худшем случае также выполняется до n раз.

Таким образом, временная сложность алгоритма составляет:

O(n²) — квадратичная сложность.

В лучшем случае (если массив уже отсортирован по модулю) внутренний цикл почти не выполняется, и сложность приближается к O(n).

**Задача 2**

Дана прямоугольная матрица размером m×n. Необходимо найти максимальный элемент его позицию (номер строки и столбца).

Запрещено использовать:
max(),
numpy
любые встроенные функции поиска

Матрица генерируется случайным образом.

Протестировать на:
матрице 2×3, матрице 1×1, матрице с отрицательными числами

Алгоритм выполняет полный перебор всех элементов матрицы.

На каждом шаге сравнивается текущий элемент с найденным максимумом.

Если элемент больше — обновляем максимум и запоминаем его координаты.

```python
def find_matrix_max(matrix):
    if len(matrix) == 0 or len(matrix[0]) == 0:
        raise ValueError("Матрица не должна быть пустой")
    
    max_value = matrix[0][0]
    max_row = 0
    max_col = 0
    
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] > max_value:
                max_value = matrix[i][j]
                max_row = i
                max_col = j
    
    return max_value, max_row, max_col
```

```python
import random

# 1. 2×3
matrix1 = [
    [1, 5, 2],
    [7, 3, 4]
]
print("Матрица 2×3:", find_matrix_max(matrix1))


# 2. 1×1
matrix2 = [[42]]
print("Матрица 1×1:", find_matrix_max(matrix2))


# 3. С отрицательными числами
matrix3 = [
    [-10, -3, -7],
    [-1, -20, -5]
]
print("Отрицательные:", find_matrix_max(matrix3))


# 4. Случайная матрица 4×5
matrix4 = [[random.randint(-50, 50) for _ in range(5)] for _ in range(4)]

print("Случайная матрица:")
for row in matrix4:
    print(row)

print("Максимум:", find_matrix_max(matrix4))
```

Используются два вложенных цикла: внешний по строкам (m) и внутренний по столбцам (n)

Каждый элемент матрицы проверяется ровно один раз.

Временная сложность: O(m × n)

Если матрица квадратная n×n: O(n²)

**Задача 3**

Необходимо определить, является ли она палиндромом. Проверку реализовать через работу с индексами.

Палиндром — это строка, которая читается одинаково слева направо и справа налево.

Условия:

Не использовать [::-1],
Не использовать reversed(),
Не использовать готовые функции сравнения с перевёрнутой строкой

Протестировать на:

пустой строке,
одном символе,
палиндроме,
непалиндроме

Алгоритм сравнивает символы попарно: первый с последним, второй с предпоследним и так далее.

Если хотя бы одна пара не совпадает — строка не является палиндромом.

Если все пары совпадают — строка является палиндромом.

```python
def is_palindrome(text):
    n = len(text)
    
    for i in range(n // 2):
        if text[i] != text[n - 1 - i]:
            return False
    
    return True
```

```python
# 1. Палиндром
print("level:", is_palindrome("level"))

# 2. Не палиндром
print("hello:", is_palindrome("hello"))

# 3. Пустая строка
print("':", is_palindrome(""))

# 4. Один символ
print("a:", is_palindrome("a"))

# 5. Чётная длина
print("abba:", is_palindrome("abba"))
```

Алгоритм выполняет один проход до середины строки длины n.

Количество сравнений примерно n / 2.

Временная сложность:
O(n)
